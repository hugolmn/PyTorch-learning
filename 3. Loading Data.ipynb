# Loading Data

```python
import torch
import torchvision
```

### Loading dataset from torchvision package

```python
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=False)
```

```python
trainset
```

### Shuffle and batch data using DataLoader

```python
trainloader = torch.utils.data.DataLoader(
    dataset=trainset,
    batch_size=4,
    shuffle=True,
    num_workers=2
)
```
